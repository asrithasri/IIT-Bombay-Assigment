```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/IIT bombay Internsip folder/AH_Monthly_Provisional_Counts_of_Deaths_by_Age_Group_and_HHS_region_for_Select_Causes_of_Death__2019-2021.csv')
df.head()
```

```python
# Part 2 Time-Wise Death Analysis: US Data

# Filter the data for the specified conditions
filtered_data = df[(df['AgeGroup'] == '15-24 years') & (df['Date Of Death Year'].isin([2019, 2020, 2021]))]

# Grouping the data by year
grouped_data = filtered_data.groupby(['Date Of Death Year', 'Date Of Death Month']).sum().reset_index()

# Plotting the line graph
colors = {2019: 'red', 2020: 'green', 2021: 'blue'}
fig, ax = plt.subplots()

for year, data in grouped_data.groupby('Date Of Death Year'):
    ax.plot(data['Date Of Death Month'], data['NaturalCause'], label=str(year), color=colors[year])

#  labels and title
ax.set_xlabel('Month')
ax.set_ylabel('Natural Cause')
ax.set_title('Time-Wise Death Analysis: US Data -> Natural Cause Deaths in US:(AgeGroup 15-24 years) ')
ax.legend(title='Year')

# Show the plot
plt.show()
```
